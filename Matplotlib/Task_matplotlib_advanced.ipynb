### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
# code here
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
zz=np.absolute(xx)-np.absolute(yy)
ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,zz,cmap='jet')
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
# code here
x=np.linspace(-10,10,10)
y=np.linspace(-10,10,10)
xx,yy=np.meshgrid(x,y)
zz=np.sin(np.sqrt( (xx**2) + (yy**2) ))
````

````python
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,zz,cmap='viridis')
fig=plt.figure()
fig.colorbar(p)
````

````python
ax=plt.subplot()
p=ax.contour(xx,yy,zz)
fig=plt.figure()
fig.colorbar(p)
````

````python
ax=plt.subplot()
p=ax.contourf(xx,yy,zz)
fig=plt.figure()
fig.colorbar(p)
````

### `Problem 10:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = tan(\log_2({x^2 + y^2})$$

````python
# code here
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
zz=np.tan(np.log2( (xx**2) + (x**2) ))
````

````python
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,zz,cmap='viridis')
fig=plt.figure()
fig.colorbar(p)
````

````python
ax=plt.subplot()
p=ax.contour(xx,yy,zz,cmap='viridis')
fig=plt.figure()
fig.colorbar(p)
````

````python
ax=plt.subplot()
p=ax.contourf(xx,yy,zz,cmap='viridis')
fig=plt.figure()
fig.colorbar(p)
````
