### Subplots

```python
fig,ax=plt.subplots()
ax.plot(batters['avg'],batters['strike_rate'],'*')
ax.set_title('Title')
ax.set_xlabel('Average')
ax.set_ylabel('Strike rate')
# fig.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[0].set_title("Avg VS Strike rate")
ax[0].set_ylabel("Strike Rate")
ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title("Avg VS Runs")
ax[1].set_xlabel("Average")
ax[1].set_ylabel("Runs")
```

```python
figu=plt.figure()
ax1=figu.add_subplot(2,1,1)
ax1.scatter(batters['avg'],batters['strike_rate'])
ax2=figu.add_subplot(2,1,2)
ax2.scatter(batters['avg'],batters['runs'])
```

### 3D Scatter Plot

```python
fig2=plt.figure()
ax11=plt.subplot(projection='3d')
ax11.scatter3D(batters['runs'],batters['avg'],batters['strike_rate'],marker='*',color='green')
ax11.set_xlabel("Runs")
ax11.set_ylabel("Average")
ax11.set_zlabel("Strike Rate")
```

### 3D Line Plot

```python
x=[0,1,5]
y=[0,10,13]
z=[0,13,20]
fig2=plt.figure()
ax11=plt.subplot(projection='3d')
ax11.plot3D(x,y,z)
```

### 3D Surface Plot

```python
xval=np.linspace(-10,10,100)
yval=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(xval,yval)
```

```python
zz=xx**2+yy**2
```

```python
fig2=plt.figure()
ax11=plt.subplot(projection='3d')
p=ax11.plot_surface(xx,yy,zz,cmap='viridis')
fig2.colorbar(p)
```

### Contour Plot

```python
fig2=plt.figure()
ax11=plt.subplot()
p=ax11.contour(xx,yy,zz,cmap='viridis')
fig2.colorbar(p)
```

```python
fig2=plt.figure()
ax11=plt.subplot()
p=ax11.contourf(xx,yy,zz,cmap='viridis')
fig2.colorbar(p)
```
