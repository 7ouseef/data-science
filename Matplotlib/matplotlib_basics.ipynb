### Types of data 
 1.Numerical data <br>
 2.Categorical data

univariate analysis -> plot on one columns <br>
bivariate analysis -> plot on two columns <br>
multivariate analysis -> plot on more than two columns <br>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2D plots
Bivariate analysis <br>
all combinations of numerical and categorical data <br>
use case - time series data

```python
price=[48000,54000,57000,49000,47000,45000] # numerical
year=[2015,2016,2017,2018,2019,2020] # categorical

plt.plot(year,price)
```

```python
batsman=pd.read_csv('./datasets/sharma-kohli.csv')
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=1)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=3)
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=1,marker='.',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=3)
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],label='Virat',color='green',linestyle='solid',linewidth=1,marker='.',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit',color='black',linestyle='dashed',linewidth=3)
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
plt.legend(loc='lower right') # loc(default)=best
```

```python
# outlier at last
price=[48000,54000,57000,49000,47000,45000,4500000] # numerical
year=[2015,2016,2017,2018,2019,2020,2021] # categorical

plt.plot(year,price)
# trimming x and y axis
plt.ylim(0,100000)
plt.xlim(2017,2021)
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=1,marker='.',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=3)
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
plt.grid()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='solid',linewidth=1,marker='.',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=3)
plt.title('V.Kohli vs R.Sharma Career Comparison')
plt.xlabel('Season of IPL')
plt.ylabel('Runs scored')
plt.grid()
plt.show() # to show graph in code editor
```

### Scatter plot
bivariate analysis <br>
numerical vs numerical <br>
use cases - finding correlation <br>

```python
x=np.linspace(-10,10,50)
y=10*x + 3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
batter=pd.read_csv('./datasets/batter.csv')
batter=batter.head(50)
batter.head(3)
```

```python
plt.scatter(batter['avg'],batter['strike_rate'],color='red',marker='.')
plt.xlabel('Average Runs')
plt.ylabel('Strike Rate')
plt.title('Average Runs VS Strike Rate')
plt.grid()
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
# s parameter in scatter
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10,color='red',marker='*')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total bill VS Tip')
plt.grid()
```

```python
plt.plot(tips['total_bill'],tips['tip'],'+') # this is faster
```

### Bar chart
categorical VS numerical(aggregate function on categorical column) <br>
use case - aggregate analysis of groups.e.g.count of categories

```python
children=[10,4,2,7,6]
color=['red','blue','green','yello','purple']
plt.bar(color,children,color='orange')
plt.xlabel('Color')
plt.ylabel('Children')
```

```python
children=[10,4,2,7,6]
color=['red','blue','green','yello','purple']
# horizontal bar chart
plt.barh(color,children,color='black')
plt.xlabel('Color')
plt.ylabel('Children')
```

```python
best_bat=pd.read_csv('./datasets/batsman_season_record.csv')
best_bat
```

```python
plt.bar(best_bat['batsman'],best_bat['2015'])
plt.bar(best_bat['batsman'],best_bat['2016'])
# overlapping grpahs so cannot see proper info
```

```python
# fixed
plt.bar(np.arange(best_bat.shape[0])-0.2,best_bat['2015'],width=0.2,label='2015')
plt.bar(np.arange(best_bat.shape[0]),best_bat['2016'],width=0.2,label='2016')
plt.bar(np.arange(best_bat.shape[0])+0.2,best_bat['2017'],width=0.2,label='2017')
plt.xticks(np.arange(best_bat.shape[0]),best_bat['batsman']) # replaces x axis indexex from 0,1,2,3,4 to batsman names
plt.legend()
plt.show()
```

```python
# problem - readibility of x or y axis
children=[10,4,2,7,6]
color=['color is red','color is blue','color is green','color is yellow','color is purple']
plt.bar(color,children)
```

```python
# solution
children=[10,4,2,7,6]
color=['color is red','color is blue','color is green','color is yellow','color is purple']
plt.bar(color,children)
plt.xticks(rotation=45)
```

```python
# stacked bar chart
plt.bar(best_bat['batsman'],best_bat['2015'],label='2015')
plt.bar(best_bat['batsman'],best_bat['2016'],bottom=best_bat['2015'],label='2016')
plt.bar(best_bat['batsman'],best_bat['2017'],bottom=best_bat['2015']+best_bat['2016'],label='2017')
plt.legend()
```

### Pie chart
categorical data <br>
use case - to find contribution of a category in data

```python
pie_d=[23,45,25,78,14]
sub=['Eng','Urdu','Sst','Maths','Science']
plt.pie(pie_d,labels=sub)
plt.show()
```

```python
gayle=pd.read_csv('./datasets/gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%')
plt.show()
# to show percentage -> autopct='%0.1f%%'
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','brown','pink','orange'])
plt.show()
```

```python
# displace a single pie
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0])
plt.show()
```
