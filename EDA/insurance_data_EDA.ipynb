# bmi - claim

```python
def bmi_fun(bmi):
    if bmi<18.5:
        return "underweight"
    elif bmi>=18.5 and bmi<=24.9:
        return "normal"
    elif bmi>=25 and bmi<=29.9:
        return "overweight"
    else:
        return "obese"
df['bmi_cat']=df['bmi'].apply(bmi_fun)
```

```python
df['bmi_cat'].value_counts()
```

```python
df['bmi_cat'].value_counts().plot(kind='bar')
```

```python
df[['bmi_cat','claim']].groupby('bmi_cat').agg([min,(lambda x: np.percentile(x,q=25)),'median'
                                                ,(lambda x: np.percentile(x,q=75)),max,'mean','std'])
# minimum claim of underweight people is the most highest among all
# despite the difference in frequency of all categories the minimum and maximum values
# for all categories are near to each other,whereas there is slight diff in median values
```

```python
df[['bmi_cat','claim']].groupby('bmi_cat').mean().sort_values('claim',ascending=False).plot(kind='bar')
```

```python
sns.boxplot(data=df,x='bmi_cat',y='claim')
plt.grid()
plt.axhline(5000)
# mean claim of obese people is highly influenced by outliers
```

```python
sns.violinplot(data=df,x='bmi_cat',y='claim')
# normal and overweight people have almost same distribution of claim
```

# bloodpressure - claim

```python
def bp_fun(bp):
    if bp<80:
        return 'optimal'
    elif bp>=80 and bp<=84:
        return 'normal'
    elif bp>=85 and bp<=89:
        return 'high normal'
    elif bp>=90 and bp<=99:
        return 'grade 1'
    elif bp>=100 and bp<=109:
        return 'grade 2'
    else:
        return 'grade 3'

df['bp_cat']=df['bloodpressure'].apply(bp_fun)
```

```python
df['bp_cat'].value_counts()
```

```python
df['bp_cat'].value_counts().plot(kind='bar')
```

```python
df[['bp_cat','claim']].groupby('bp_cat').agg([min,(lambda x: np.percentile(x,q=25)),'median'
                                                ,(lambda x: np.percentile(x,q=75)),max,'mean','std'])

# grdae 2 and grade 3 BP patients tend to reuqest higher claim amount as compared to others
```

```python
sns.boxplot(data=df,x='bp_cat',y='claim')

# grade 2 have more outliers than grade 3 so we can be sure that grade 3 patients will request higher claim amount
```

```python
sns.violinplot(data=df,x='bp_cat',y='claim')

# grade 1,high normal and normal patients have almost same distribution but grade 1 mean is highly influenced by outliers
```
