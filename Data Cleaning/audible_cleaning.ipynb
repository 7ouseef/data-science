```python
# should be in timedelta or represented in number of minutes(contains hr,mins,hrs,min)
def time(lis):
    if lis[0]=='Less than 1 minute':
        return 0
    if len(lis)>1:
        hr=lis[0].rstrip().rstrip('hrs').rstrip('hr')
        mi=lis[1].rstrip('mins').rstrip('min')
        return (int(hr)*60)+int(mi)
    else:
        if lis[0].find('hr')>-1:
            return (int(lis[0].rstrip().rstrip('mins').rstrip('min').rstrip('hr').rstrip('hrs')))*60
        else:
            return int(lis[0].rstrip().rstrip('mins').rstrip('min').rstrip('hr').rstrip('hrs'))
df['time']=df['time'].str.split('and').apply(time)
```

```python
import re
```

```python
# narrator,author: first and last name are not separated by space
def separate_name(name_list):
    ans=[]
    for name in name_list:
        if name=='unknown':
            return name_list
        all_alph=re.findall('[A-Z]',name);
        for alph in all_alph[1:]:
            i=name.find(alph,1)
            name=name[:i] + " " + name[i:]
        ans.append(name)
    return ans
df['author']=df['author'].apply(separate_name)
df['narrator']=df['narrator'].apply(separate_name)
```
