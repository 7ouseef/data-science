### Filtering data

### Matches Data

```python
matches
```

```python
# find all final winners

matches[matches['MatchNumber']=='Final'][['Season','WinningTeam']]
```

```python
# how many super over finishes have occured

(matches['SuperOver']=='Y').sum()
```

```python
# how many matches has csk won in kolkata

matches[(matches['WinningTeam']=='Chennai Super Kings') & (matches['City']=='Kolkata') ]
```

```python
i = (matches['WinningTeam']=='Chennai Super Kings') & (matches['City']=='Kolkata')
i.sum()
```

```python
# TOSS WINNER IS MATCH WINNER IN PERCENTAGE

x=matches[matches['TossWinner']==matches['WinningTeam']].shape[0]
(x/matches.shape[0])*100
```
