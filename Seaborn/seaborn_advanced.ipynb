```python
import numpy as np
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
tips.head(1)
```

```python
iris.head(1)
```

## Categorical Plots
### Categorical Scatter Plots
- strip plot
- swarm plot
### Categorical Distribution Plots
- box plot
- violin plot
### Categorical Estimate Plots(for central tendency)
- bar plot
- point plot
- count plot

```python
# strip plot -> between a categorical and a numerical column
sns.stripplot(data=tips,x='day',y='total_bill',hue='day',jitter=False) # axis level function
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',hue='day',jitter=False,kind='strip')
```

```python
# jitter
sns.catplot(data=tips,x='day',y='total_bill',hue='day',jitter=0.2,kind='strip')
```

```python
# swarm plot -> represents data in a better way by running an alogortihm internally
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
# better for small datasets
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

```python
# axes level function
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
# figure level function
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',hue='sex')
```

```python
sns.boxplot(data=tips,y='total_bill')
```

```python
# violin plot -> boxplot+kdeplot
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex',split=True)
```

```python
# bar plot -> aggregted value of a numerical and categorical columns
sns.barplot(data=tips,x='sex',y='total_bill')
```

```python
# black line is confidence interval
sns.barplot(data=tips,x='day',y='total_bill',errorbar=None,hue='smoker')
```

```python
sns.catplot(kind='bar',data=tips,x='day',y='total_bill',errorbar=None,estimator=np.min)
```

```python
# point plot -> bar plot without bars to show the difference
sns.pointplot(data=tips,x='day',y='total_bill',hue='smoker',errorbar=None)
```

```python
sns.pointplot(data=tips,x='sex',y='total_bill')
```

```python
# count plot -> similar to histogram which calculate frequency
sns.countplot(data=tips,x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
# facet plot
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

## Regression plot

```python
sns.regplot(data=tips,x='total_bill',y='tip') # cannot use hue parameter
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# resid plot -> plot error points of regression
sns.residplot(data=tips,x='total_bill',y='tip')
```

## Facet Grid
- Another way to create facet plot is to use facet grid

```python
g=sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

## Pair plot

```python
iris.head(1)
```

```python
sns.pairplot(data=iris)
# for same columns histogram is plotted 
# the grid is symmetrical so above and below of diagonal is same
```

```python
sns.pairplot(data=iris,hue='species')
```

## Pair Grid
- can use any kind of plot on diagonal or off diagonal

```python
sns.PairGrid(data=iris).map(sns.scatterplot)
```

```python
gr=sns.PairGrid(data=iris)
gr.map_diag(sns.boxplot)
gr.map_offdiag(sns.histplot)
```

```python
gr=sns.PairGrid(data=iris)
gr.map_diag(sns.boxplot)
gr.map_upper(sns.histplot)
gr.map_lower(sns.kdeplot)
```

```python
gr=sns.PairGrid(data=iris,vars=['sepal_width','petal_length'])
gr.map_diag(sns.boxplot)
gr.map_lower(sns.histplot)
gr.map_upper(sns.kdeplot)
```

## Joint plot

```python
# also plot histogram on x and y columns
sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

## Joint Grid

```python
sns.JointGrid(data=tips,x='total_bill',y='tip').plot(sns.scatterplot,sns.violinplot)
```
